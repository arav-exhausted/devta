```python
def distance_formula(lat1,lat2,long1,long2):
    R = 6371

    lat1=np.radians(lat1)
    lat2=np.radians(lat2)
    long1=np.radians(long1)
    long2=np.radians(long2)

    dlat = lat2 - lat1
    dlong = long2-long1

    a = (
        np.sin(dlat / 2) ** 2
        + np.cos(lat1) * np.cos(lat2) * np.sin(dlong / 2) ** 2
    )

    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    return R * c
```

```python
df['Distance_Km'] = distance_formula(
    df["Restaurant_latitude"],
    df['Delivery_location_latitude'],
    df["Restaurant_longitude"],
    df['Delivery_location_longitude']
)
```

## Feature Engineering

```python
def features(df):
    df['Peak_hour'] = (df['Order_time'].between(12,14) | df['Order_time'].between(19,21)).astype(int)
    df['Long_Distance'] = (df['Distance_Km']>10).astype(int)
    df['Late_Night_Order'] = ((df['Order_time']>=22 )| (df['Order_time']<=5))
    df['Highly_Rated_Driver'] = (df['Delivery_person_Ratings']>=4.5).astype(int)
    df["Busy_Rider"] = (df["multiple_deliveries"] >= 2).astype(int)

    return df


df = features(df)


    
```

### Split

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Time_taken(min)'])
y = df['Time_taken(min)']

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.3)
```

### Preprocessing

```python
numerical_cols = [
    'Delivery_person_Age',
    'Delivery_person_Ratings',
    'Vehicle_condition',
    'multiple_deliveries',
    'Distance_Km',
    'Order_time',
    'Is_Weekend',
    'Peak_hour',
    'Long_Distance',
    'Late_Night_Order',
    'Highly_Rated_Driver',
    'Busy_Rider'
]

for col in numerical_cols:
    median = X_train[col].median()

    X_train[col] = X_train[col].fillna(median)
    X_test[col] = X_test[col].fillna(median)
```

```python
for col in categorial_cols:
    mode = X_train[col].mode()[0]

    X_train[col] = X_train[col].fillna(mode)
    X_test[col] = X_test[col].fillna(mode)
```

```python
print(X_train.isnull().sum())
print(X_test.isnull().sum())
```

```python
X_train = pd.get_dummies(
    X_train,
    columns=categorial_cols,
    drop_first=True,
    dtype=int
)

X_test = pd.get_dummies(
    X_test,
    columns=categorial_cols,
    drop_first=True,
    dtype=int
)

X_train, X_test = X_train.align(
    X_test,
    join="left",
    axis=1,
    fill_value=0
)
```

## Model training


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error,mean_absolute_error,r2_score
```

```python

linearReg = LinearRegression()

linearReg.fit(X_test,y_test)

y_pred_LinReg = linearReg.predict(X_test)

print("RMSE", root_mean_squared_error(y_test,y_pred_LinReg))
print("MAE", mean_absolute_error(y_test,y_pred_LinReg))
print("R2 Score", r2_score(y_test,y_pred_LinReg))
```

```python
from sklearn.ensemble import RandomForestRegressor

randomforest = RandomForestRegressor()

randomforest.fit(X_train,y_train)

y_pred_randomforest = randomforest.predict(X_test)

print("RMSE", root_mean_squared_error(y_test,y_pred_randomforest))
print("MAE", mean_absolute_error(y_test,y_pred_randomforest))
print("R2 Score", r2_score(y_test,y_pred_randomforest))
```

```python
from xgboost import XGBRegressor

xgb = XGBRegressor()

xgb.fit(X_train,y_train)


y_pred_xgb = xgb.predict(X_test)


print("RMSE", root_mean_squared_error(y_test,y_pred_xgb))
print("MAE", mean_absolute_error(y_test,y_pred_xgb))
print("R2 Score", r2_score(y_test,y_pred_xgb))
```

```python
from lightgbm import LGBMRegressor

lgb = LGBMRegressor()

lgb.fit(X_train,y_train)


y_pred_lgb = lgb.predict(X_test)


print("RMSE", root_mean_squared_error(y_test,y_pred_lgb))
print("MAE", mean_absolute_error(y_test,y_pred_lgb))
print("R2 Score", r2_score(y_test,y_pred_lgb))
```

```python
from sklearn.model_selection import RandomizedSearchCV


parameters = {
    "n_estimators": [100, 200, 400, 600],
    "learning_rate": [0.03, 0.05, 0.1],
    "max_depth": [3, 5, 7, 9],
    "min_child_weight": [1, 3, 5],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0],
    "gamma": [0, 0.1, 0.3],
    "reg_alpha": [0, 0.1, 0.5],
    "reg_lambda": [1, 2, 5]
}

xgb = XGBRegressor(
    objective="reg:squarederror",
    random_state=42,
    n_jobs=-1,
    tree_method="hist",      
    verbosity=0
)


random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=parameters,
    n_iter=20,
    cv=3,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1,
    random_state=42,
    verbose=2
)


random_search.fit(X_train, y_train)


best_model = random_search.best_estimator_


y_pred = best_model.predict(X_test)

print(random_search.best_params_)
print("=" * 50)

print(f"RMSE : {root_mean_squared_error(y_test, y_pred):.4f}")
print(f"MAE  : {mean_absolute_error(y_test, y_pred):.4f}")
print(f"R²   : {r2_score(y_test, y_pred):.4f}")
```
